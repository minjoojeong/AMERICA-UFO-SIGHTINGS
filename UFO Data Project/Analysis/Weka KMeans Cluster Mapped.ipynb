```python
#import stuff
import folium 
from folium.plugins import StripePattern
import geopandas as gpd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pysal as ps
from sklearn import cluster
from sklearn.preprocessing import scale
import numpy as np

url = (
    "https://raw.githubusercontent.com/python-visualization/folium/master/examples/data"
)
state_geo = f"{url}/us-states.json"

# We read the file and print it.
geoJSON_df = gpd.read_file(state_geo)

# Next we grab the states and put them in a list and check the length.
geoJSON_states = list(geoJSON_df.id.values)
len(geoJSON_states)

sns.set(style="whitegrid")
#clusters are from Weka, using KMeans clustering
#INITIAL POINTS SET BY WEKA
#12 clusters for each region: Midwest, South, Northeast, West
ufo = pd.read_csv('randomUFO.csv')
ufo.head()
```

```python
initialUFO = folium.Map(location=[48, -102], zoom_start = 4)
folium.TileLayer('cartodbpositron').add_to(initialUFO)

for index,row in ufo.iterrows():
    lat = row['latitude']
    lon = row['longitude']
    name = str(row['shape'])
    
    folium.Circle(
        radius=(row['duration']*50),
        location=[lat,lon],
        color='red',
        opacity=0.4,
        fill=False,
    ).add_to(initialUFO)
    
    newMarker = folium.Marker(
        [float(lat),float(lon)],
        popup=name,
        icon=folium.DivIcon(html=f"""<div style="font-family: courier new; color: black">{row['shape']}</div>""")
    )
    newMarker.add_to(initialUFO)
    
initialUFO
#random set of UFO sightings
```

```python
#read csv file of clustered ufo points
#kmeans clustering done on weka for each regional data set 
#4 regions, 7 clusters each, 28 total
kmeans = pd.read_csv('cluster_ufo.csv')
#kmeans.head()
kmeans.groupby(['shape'])['shape'].count()
```

```python
shape_color = {
  "circle": 'orange',
  "fireball": 'blue',
  "light": 'red',
    'other':'yellow',
    'oval':'green',
    'sphere':'magenta',
    'triangle':'purple'
}

clusterUFO = folium.Map(location=[48, -102], zoom_start = 4)
folium.TileLayer('cartodbpositron').add_to(clusterUFO)

for index,row in kmeans.iterrows():
    lat = row['latitude']
    lon = row['longitude']
    name = str(row['shape'])
    
    folium.Circle(
        radius=(row['duration']*100),
        location=[lat,lon],
        color=shape_color[row['shape']],
        opacity=0.4,
        fill=False,
    ).add_to(clusterUFO)
    
    newMarker = folium.Marker(
        [float(lat),float(lon)],
        popup=name,
        icon=folium.DivIcon(html=f"""<div style="font-family: courier new; color: black">{row['shape']}</div>""")
    )
    newMarker.add_to(clusterUFO)
    
clusterUFO
#UFO sightings clustered
```
