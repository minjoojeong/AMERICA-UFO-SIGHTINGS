```python
import folium
import pandas as pd
#sample map of 100 data points 
#dataset is too large to map efficiently
ufoFile = pd.read_csv('mapufo.csv')
smallUFO = ufoFile.head(100)
mapUFO = folium.Map(location=[48, -102], zoom_start = 3)


for index,row in smallUFO.iterrows():
    lat = row['city_latitude']
    lon = row['city_longitude']
    name = str(index)
    newMarker = folium.Marker([float(lat),float(lon)],popup=name)
    newMarker.add_to(mapUFO)
    
mapUFO
```

```python
#import stuff
import pandas as pd
import folium 
from folium.plugins import StripePattern
import geopandas as gpd
import numpy as np
url = (
    "https://raw.githubusercontent.com/python-visualization/folium/master/examples/data"
)
state_geo = f"{url}/us-states.json"

ufo_state = pd.read_csv('mapufo.csv')
ufo_state = ufo_state[['state','date_time','shape']]
# We read the file and print it.
geoJSON_df = gpd.read_file(state_geo)

# Next we grab the states and put them in a list and check the length.
geoJSON_states = list(geoJSON_df.id.values)
len(geoJSON_states)
#stateMap = folium.Map(location=[48, -102], zoom_start=3)
#stateMap
```

```python
# we rename the column from id to state in the geoJSON_df so we can merge the two data frames.
geoJSON_df = geoJSON_df.rename(columns = {"id":"state"})
# Next we merge our sample data (df) and the geoJSON data frame on the key id.
final_df = geoJSON_df.merge(ufo_state, on = "state")
final_df.head()
```

```python
state_group = final_df.groupby("state")["shape"].count()
state_group.head()
state_group_geo = geoJSON_df.merge(state_group, on = "state")
state_group_geo.head()
```

```python

# Initialize folium map.
sample_map = folium.Map(location=[48, -102], zoom_start=3)
sample_map
# Set up Choropleth map
folium.Choropleth(
geo_data=state_group_geo,
data=state_group_geo,
columns=['state','shape'],
key_on="feature.properties.state",
fill_color='YlGnBu',
fill_opacity=1,
line_opacity=0.2,
legend_name="shape",
smooth_factor=0,
Highlight= True,
line_color = "#0000",
name = "Shape",
show=False,
overlay=True,
nan_fill_color = "White"
).add_to(sample_map)

sample_map
```
